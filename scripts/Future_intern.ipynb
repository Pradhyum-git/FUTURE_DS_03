### Model Building 

```python
# It is a Regression Problem 
```

```python
x=df.drop('course_recommendation',axis=1)
y=df['course_recommendation']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=25)
```

```python
print("Input training Data:",x_train.shape)
print("Training Output:",y_train.shape)
print("Test Input Size:",x_test.shape)
print("Test Output:",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
def model_evaluation(model,x_train,y_train,x_test,y_test):
    model=model
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    R_squared=r2_score(y_test,y_pred)
    mse=mean_squared_error(y_test,y_pred)
    rmse=np.sqrt(mse)
    n=x_test.shape[0]
    k=x_test.shape[1]
    Adj_R_squared = 1 - (1 - R_squared) * ((n - 1) / (n - k - 1))
    return mse,rmse,R_squared,Adj_R_squared
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
mse,rmse,r2,adj_r2=model_evaluation(lr,x_train,y_train,x_test,y_test)
print("Mean Squared Error:",mse)
print("Root Mean Squared Error:",rmse)
print("R2 Score:",r2)
print("Adjusted R2 Score:",adj_r2)
```
